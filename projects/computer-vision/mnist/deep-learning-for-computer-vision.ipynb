# Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
from keras import layers
from keras.models import Model
import os
```

# Dense Model

```python
def DenseModel(dim_1, dim_2):
    inputs = layers.Input(shape = (dim_1, dim_2))
    
    outputs = layers.Dense(10, activation='softmax')(inputs)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model
```

```python
DenseModel(28, 28).summary()
```
